## 1. Raw Data Ingestion

Load the raw CSV and take an initial look.

```python
df_raw = pd.read_csv(RAW_PATH)
print(f'Shape : {df_raw.shape}')
print(f'Columns: {list(df_raw.columns)}')
df_raw.head(3)
```

```python
# Data-type snapshot: note quantity and unit_price read as 'object' due to
# the non-numeric garbage tokens injected to simulate Spark PERMISSIVE-mode
# corrupt records. The PySpark ingestion stage routes these rows out first.
df_raw.dtypes
```

## 2. Data Quality Assessment

Replication of the Spark pipeline's ingestion + cleaning logic in pandas so this notebook can show real numbers without a running Spark cluster.

```python
# Route structurally corrupt records (non-numeric typed columns)
qty_num   = pd.to_numeric(df_raw['quantity'],   errors='coerce')
price_num = pd.to_numeric(df_raw['unit_price'],  errors='coerce')
corrupt_mask = qty_num.isna() | price_num.isna()

print(f'Malformed records routed at ingestion : {corrupt_mask.sum():>6,}')
print(f'Usable rows after routing             : {(~corrupt_mask).sum():>6,}')
```

```python
df = df_raw.loc[~corrupt_mask].copy()
df['quantity']   = qty_num.loc[~corrupt_mask].astype(int)
df['unit_price'] = price_num.loc[~corrupt_mask].astype(float)

issues = {
    'Missing customer_id'       : df['customer_id'].isna().sum(),
    'Missing payment_method'    : df['payment_method'].isna().sum(),
    'Missing category'          : df['category'].isna().sum(),
    'Missing city'              : df['city'].isna().sum(),
    'Negative quantity'         : (df['quantity'] < 0).sum(),
    'Duplicate rows (exact)'    : df.duplicated(keep=False).sum(),
    'Duplicate transaction_ids' : df.duplicated(subset=['transaction_id'], keep=False).sum(),
}
for k, v in issues.items():
    print(f'{k:<30}: {v:>6,}')
```

```python
# ── Load pre-computed report CSVs so the chart cells below are self-contained ──
# These CSVs are committed to the repo (reports/) and produced by the PySpark
# pipeline.  Loading them here means the notebook runs without a live Spark
# session or MySQL connection.
revenue_by_category = pd.read_csv('../reports/category_sales.csv')
revenue_by_state    = pd.read_csv('../reports/state_sales.csv')
monthly_trend       = pd.read_csv('../reports/monthly_sales.csv')
top_products        = pd.read_csv('../reports/top_products.csv')
top_cities          = pd.read_csv('../reports/city_sales.csv')
print('Report CSVs loaded.')
print(f'  Categories : {len(revenue_by_category)}')
print(f'  States     : {len(revenue_by_state)}')
print(f'  Months     : {len(monthly_trend)}')
print(f'  Top prods  : {len(top_products)}')
print(f'  Top cities : {len(top_cities)}')
```

## 6. Revenue & Profit by Category

```python
fig, ax = plt.subplots(figsize=(9, 5))
cats_df = revenue_by_category.sort_values('total_revenue', ascending=True)
ax.barh(cats_df['category'], cats_df['total_revenue'], color=TEAL, label='Revenue')
ax.barh(cats_df['category'], cats_df['total_profit'],  color=TEAL_LIGHT, label='Profit (modeled)')
ax.set_xlabel('₹')
ax.set_title('Revenue & Modeled Profit by Category')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))
ax.legend(frameon=False, loc='lower right')
plt.tight_layout()
plt.show()
```

## 7. Monthly Revenue Trend

```python
fig, ax = plt.subplots(figsize=(10, 4.5))
ax.plot(monthly_trend['order_year_month'], monthly_trend['total_revenue'],
        color=TEAL, marker='o', markersize=3, linewidth=1.5)
ax.set_title('Monthly Revenue Trend (2024–2025)')
ax.set_ylabel('Revenue (₹)')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))
tick_positions = list(range(0, len(monthly_trend), 3))
ax.set_xticks([monthly_trend['order_year_month'].iloc[i] for i in tick_positions])
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

## 8. Top 10 Products by Revenue

```python
fig, ax = plt.subplots(figsize=(9, 5))
prods_df = top_products.sort_values('total_revenue', ascending=True)
ax.barh(prods_df['product_id'], prods_df['total_revenue'], color=TEAL)
ax.set_xlabel('Revenue (₹)')
ax.set_title('Top 10 Products by Revenue')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1e3:.0f}K'))
plt.tight_layout()
plt.show()
```

## 9. Geographic Breakdown

```python
fig, ax = plt.subplots(figsize=(9, 5))
cities_df = top_cities.sort_values('total_revenue', ascending=True)
labels = cities_df['city'] + ' (' + cities_df['state'] + ')'
ax.barh(labels, cities_df['total_revenue'], color=TEAL)
ax.set_xlabel('Revenue (₹)')
ax.set_title('Top 10 Cities by Revenue')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))
plt.tight_layout()
plt.show()
```

## 10. Key Findings

- **Electronics** is the top revenue category at ₹1,463,142,567, but **Beauty & Personal Care** has the highest modeled profit margin (40%).
- **Gujarat, Maharashtra, and Uttar Pradesh** are the top-3 states, together accounting for over 45% of total revenue.
- Average delivery time of **5.51 days** (after excluding 977 date-integrity anomalies).
- Data quality: **99%** revenue consistency and **99%** date integrity across 97,961 validated orders.
